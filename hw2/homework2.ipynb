```python
import pandas as pd
```

```python
combine_df = pd.read_pickle('tweets.pkl')
```

```python
combine_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vectorizer = CountVectorizer(ngram_range=[1,3], analyzer='word', \
                                   max_features=1000, max_df=0.9, stop_words='english')
```

```python
stemmed_bag = count_vectorizer.fit_transform(combine_df["tweet_stemmed"].str.join(' '))
```

```python
stemmed_feature_names = count_vectorizer.get_feature_names()
pd.DataFrame(stemmed_bag.toarray(), columns=stemmed_feature_names)
```

```python
lemmatized_bag = count_vectorizer.fit_transform(combine_df["tweet_lemmatized"].str.join(' '))
```

```python
lemmatized_feature_names = count_vectorizer.get_feature_names()
pd.DataFrame(lemmatized_bag.toarray(), columns=lemmatized_feature_names)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer(ngram_range=[1,3], max_df=0.9, max_features=1000, stop_words='english')
```

```python
stemmed_tfidf = tfidf_vectorizer.fit_transform(combine_df["tweet_stemmed"].str.join(' '))
```

```python
stemmed_tfidf_features = tfidf_vectorizer.get_feature_names()
pd.DataFrame(stemmed_tfidf.toarray(), columns=stemmed_tfidf_features)
```

```python
lemmatized_tfidf = tfidf_vectorizer.fit_transform(combine_df["tweet_lemmatized"].str.join(' '))
```

```python
lemmatized_tfidf_features = tfidf_vectorizer.get_feature_names()
pd.DataFrame(lemmatized_tfidf.toarray(), columns=lemmatized_tfidf_features)
```
